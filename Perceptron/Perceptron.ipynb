```python
from numpy import dot
import pandas as pd
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
```

```python
class Perceptron(object):

    def __init__(self, in_size, bias, learning_rate=0.01, epochs=50):
        self.W = np.zeros(in_size)
        self.bias = bias
        self.learning_rate = learning_rate
        self.epochs = epochs
        
    def summation_step(self, X):
        #print(X.shape)
        #print(self.W.T.shape)
        res = dot(self.W.T, X) + self.bias
        a = self.activation_step(res)
        return a

    def activation_step(self, res):
        if res >= 0:
            return 1
        return 0

    def fit(self, X, y):
        for _ in range(self.epochs):
            for i in range(X.shape[0]):
                yhat = self.summation_step(X[i])
                delta = self.learning_rate * (y[i] - yhat)
                #print(y[i])
                #print(yhat)
                #print(delta)
                #print(delta.shape)
                #print(X.shape)
                #print(W.shape)
                self.W += delta * X[i]
                self.bias += self.bias * self.learning_rate

    def predict(self, X):
        #print (self.W.shape)
        res = np.dot(X, self.W) + self.bias
        #return np.where(res >= 0.0, 1, -1)
        return np.where(res >= 0.0, 1, 0)

```

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', header=None)

# setosa and versicolor
y = df.iloc[0:100, 4].values

y = np.where(y == 'Iris-setosa', 0, 1)

# sepal length and petal length
X = df.iloc[0:100, [0,2]].values

```

```python


epochs=10
learning_rate=0.1

#bias = np.random.rand(1)[0]
bias = 0

tron = Perceptron(2, 0, learning_rate, epochs)
tron.fit(X, y)
#print('Weights: %s' % ppn.w_)

plot_decision_regions(X, y, clf=tron)
plt.title('Perceptron')
plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.show()

'''
plt.plot(range(1, len(ppn.errors_)+1), ppn.errors_, marker='o')
plt.xlabel('Iterations')
plt.ylabel('Misclassifications')
plt.show()
'''
```

```python
sonar_df = pd.read_csv('data/sonar.all-data.txt', header=None)
```

```python
sonar_df.head()
sonar_df = sonar_df.apply(pd.to_numeric, errors='ignore')
sonar_df.dtypes
```

```python
##Need to add errors to the perceptron to be able to say where we are going wrong esp for high dimensional feature data
class Perceptron(object):

    def __init__(self, in_size, bias, learning_rate=0.01, epochs=50):
        self.W = np.zeros(in_size)
        self.bias = bias
        self.learning_rate = learning_rate
        self.epochs = epochs
        
    def summation_step(self, X):
        res = dot(self.W.T, X) + self.bias
        a = self.activation_step(res)
        return a

    def activation_step(self, res):
        if res >= 0:
            return 1
        return 0

    def fit(self, X, y):
        self.error = []
        for _ in range(self.epochs):
            errors = 0
            for i in range(X.shape[0]):
                yhat = self.summation_step(X[i])
                delta = self.learning_rate * (y[i] - yhat)
                self.W += delta * X[i]
                self.bias += self.bias * self.learning_rate
                errors += int(delta != 0.0)
            self.error.append(errors)

    def predict(self, X):
        #print (self.W.shape)
        res = np.dot(X, self.W) + self.bias
        return np.where(res >= 0.0, 1, 0)
    
    
```

```python

y = sonar_df.iloc[:,60].values
X = sonar_df.iloc[:,:59].values

y = np.where(y == 'R', 1, 0)

epochs=25
learning_rate=0.01


bias = 0

tron = Perceptron(59, 0, learning_rate, epochs)
tron.fit(X, y)

plt.plot(range(1, len(tron.error)+1), tron.error, marker='o')
plt.xlabel('Iterations')
plt.ylabel('Misclassifications')
plt.show()

plt.show()
```

```python
def accuracy():
    num_obs = X.shape[0]
    scores = list()
    for i in range(len(tron.error)):
        accuracy = num_obs - tron.error[i]
        scores.append((accuracy / num_obs) * 100.0)
    return scores

scores = accuracy()
```

```python
print('Scores: %s' % scores)
print('Mean Accuracy: %.3f%%' % (sum(scores)/float(len(scores))))
```
